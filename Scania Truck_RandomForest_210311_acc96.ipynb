### 1. Data Processing

    - histogram에 대한 처리: 대표값으로 대체
        - 평균, 분산
    - NA 값에 대한 처리: columnwise 적용. pos null 30%이상 -> pos median, 나머지 -> 전체의 median 
    - class imbalance: oversampling (SMOTE)
    - feature selection: drop over 40% na, 2장 내용 참고, logistic 회귀
    - feature reduction: PCA
    - columnwise normalizing
    

```python
def replaceHisto(data):
    prefixes = Counter([col.split("_")[0] for col in data.columns])
    histo_cols = dict()
    
    for pre, cnt in prefixes.items():
        if cnt>1:
            histo_cols[pre] = [pre+"_"+str(format(num, '03d')) for num in range(cnt)]
    
    for prefix, cols in histo_cols.items():
        data[prefix+"_median"] = data[cols].median(axis=1)
        data[prefix+"_mean"] = data[cols].mean(axis=1)
        data[prefix+"_var"] = data[cols].var(axis=1)
        data.drop(cols, inplace=True, axis=1)

    print(f"Changed historgram cols to mean, median, variation\n")     
    return data

def processNA(data, mode='Train', neg_median=None, to_drop=None):
    if mode=='Train':
        pos_data = data[data["class"]==1] 
        neg_data = data[data["class"]==0] 

        pos_df = pos_data.isnull().sum().to_frame(name="pos_null_cnt")
        neg_df = neg_data.isnull().sum().to_frame(name="neg_null_cnt")
        null_check = pos_df.join(neg_df)
        null_check["total_null"] = null_check["pos_null_cnt"] + null_check["neg_null_cnt"]
        null_check.insert(1, "pos_cnt", len(pos_data))
        null_check.insert(3, "neg_cnt", len(neg_data))

        null_check["pos_null_ratio"] = null_check.pos_null_cnt/null_check.pos_cnt
        null_check["neg_null_ratio"] = null_check.neg_null_cnt/null_check.neg_cnt
        null_check["total_null_ratio"] = (null_check.pos_null_cnt + null_check.neg_null_cnt)/len(data)

        # drop cols w/ too many na's
        to_drop = null_check[null_check.total_null_ratio>=0.4]
        to_drop = set(to_drop.index)
    
        # neg_median
        neg_median = neg_data.median(axis=0).to_dict()
    else:
        if neg_median is None or to_drop is None: 
            print("\n:::ERROR:::NEG_MEDIAN or to_drop cannot be None in the Test mode")
            return -4
    
    data.drop(to_drop, inplace=True, axis=1)   
    data.fillna(value=neg_median, inplace=True)
    
    print(f"Dropped {len(to_drop)} columns: {','.join(to_drop)}")
    print(f"Replaced with NEG_MEDIAN\n")
    
    return data, neg_median, to_drop

def standardize(data, mode='Train', scaler = None):
    if mode=='Train': 
        scaler = StandardScaler()
        scaler.fit(data)
    else: 
        if scaler is None: 
            print("\n:::ERROR:::scaler cannot be None in the Test mode")
            return -3

    scaled_data = scaler.transform(data)     
    scaled_data = pd.DataFrame(scaled_data, columns=data.columns, index=data.index)
    
    print(f"Standardization finished\n")
    print(scaled_data.describe().applymap("{0:.2f}".format))
    return scaled_data, scaler

def dimReduction(data, mode='Train', pca_fitted = None):
    if mode=='Train': 
        pca = PCA()
        pca.fit(data)

        cumsum_ex_var = np.cumsum(pca.explained_variance_ratio_)
        dim = np.argmax(cumsum_ex_var >= 0.95) + 1
        print("number of dimensions: " + str(dim))

        plt.figure(figsize=(5, 5))
        plt.xlabel('Number of components')
        plt.ylabel('cumulative sum of explained variance')
        plt.plot(range(1, cumsum_ex_var.shape[0] + 1), cumsum_ex_var)

        pca = PCA(n_components=dim) # 
        pca.fit(data)

        print("explained variance ratio: " + str(sum(pca.explained_variance_ratio_))+'\n')

    else : 
        if pca_fitted is None:
            print("\n:::ERROR:::pca cannot be None in the Test mode\n")
            return -4
        pca = pca_fitted

    data_reduced = pd.DataFrame(pca.transform(data))
    print('PCA finished\n')
    return data_reduced, pca

def featureSelection(X_data, y_data, mode='Train', to_drop=None):
    # if already fitted, input columns to drop at to_drop pamam
    if mode=='Train':
        forest = ExtraTreesClassifier(max_depth=6, 
                                      min_samples_leaf=3,
                                      min_samples_split=10,
                                      n_estimators=100, random_state=42)
        forest.fit(X_data, y_data)
        feature_importances = forest.feature_importances_

        #변수중요도 barplot
        ft_importances = pd.Series(feature_importances, index = X_data.columns)
        ft_importances = ft_importances.sort_values(ascending = False)

        plt.figure(figsize = (100,100))
        plt.title("Feature Importance")
        plt.barh(ft_importances.index, ft_importances.values)      
#         plt.plot([0,0.0001], [len(ft_importances.index),0.0001], 'y-')

#         for x,y in zip(ft_importances.index, ft_importances.values):
#             label = "{:.4f}".format(y)

#             plt.annotate(label, # this is the text
#                          (x,y), # this is the point to label
#                          textcoords="offset points", # how to position the text
#                          xytext=(0,3), # distance from text to points (x,y)
#                          ha='center') # horizontal alignment can be left, right or center
        plt.show()

        # 제거할 feature 선택
        THRESHOLD = 0.0001
        to_drop = ft_importances[ft_importances<=THRESHOLD].index

    else: 
        if to_drop is None: 
            print("\n:::ERROR:::to_drop cannot be None if fitted")
            return -5

    X_data.drop(columns = to_drop, inplace=True, axis=1)
    print(f"Dropped feature: {', '.join(to_drop)}"+'\n')

    return X_data, to_drop

def classBalancing(X_data, y_data, rndn_state=42):
    print('Original dataset shape: %s' % Counter(y_data))

    sm = SMOTE(random_state=rndn_state)
    X_res, y_res = sm.fit_resample(X_data, y_data)
    print('Resampled dataset shape: %s' % Counter(y_res))

    scaler = StandardScaler()
    scaled_X = scaler.fit_transform(X_res)
    scaled_X = pd.DataFrame(scaled_X, columns=X_res.columns, index=X_res.index)

    print(scaled_X.describe())
    print('Oversampling finished\n')
    
    return scaled_X, y_res, scaler
```

```python
# load file
src_data = pd.read_csv("./Train_data.csv", index_col=0)
src_data['class'] = src_data['class'].replace('neg', 0)
src_data['class'] = src_data['class'].replace('pos', 1)

# Common Processing - 1
src_data = src_data.replace('na', np.nan) # change 'na' ro np.nan
src_data = src_data.apply(pd.to_numeric) # change object to int or float
src_data.rename(columns = {'am_0': 'am_000'}, inplace = True) # fix col name
src_data, NEG_MEDIAN, TODROP_1 = processNA(data=src_data, mode='Train', to_drop=None) # 1) Drop col w/ too much NA's. 2) Replace na with NEG_MEDIAN

X_data, y_data = src_data.iloc[:, 1:], src_data['class']

# Common Processing - 2
X_data = replaceHisto(X_data) # Replace Histogram
X_data, SCALER_1 = standardize(data=X_data, mode='Train', scaler=None) # Standardizing
# X_data, TODROP_2 = featureSelection(X_data=X_data, y_data=y_data, mode='Train', to_drop=None) # Feature Selection using Random Forest
X_data, PCAFITTED = dimReduction(data=X_data, mode='Train', pca_fitted=None) # PCA dim. reduction

# data split
X_trainval, X_test, y_trainval, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=44, stratify=data["class"])
X_train, X_val, y_train, y_val = train_test_split(X_trainval, y_trainval, test_size=0.2, random_state=44, stratify=y_trainval)

# Train only processing - class balancing using SMOTE
X_train, y_train, SCALER_2 = classBalancing(X_data=X_train, y_data=y_train) 
X_val, _ = standardize(data=X_val, mode='Test', scaler=SCALER_2) # Standardizing w/ balanced train data
X_test, _ = standardize(data=X_test, mode='Test', scaler=SCALER_2) # Standardizing w/ balanced train data
```
